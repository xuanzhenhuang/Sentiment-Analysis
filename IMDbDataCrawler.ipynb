## Step2 保存数据函数

```python
def save_data_to_excel(data, file_path):
    import pandas as pd
    df = pd.DataFrame(data, columns=[
        "img", "title", "year", "time", "rank","rate", "count", "url", "id"
    ])
    if os.path.exists(file_path):
        existing_df = pd.read_excel(file_path)
        df = pd.concat([existing_df, df], ignore_index=True)
    df.to_excel(file_path, index=False)
    print(f"数据已保存到 {file_path}")
```

## Step4 获取电影信息函数

```python
def getInfo(tab):
    current_dir = os.getcwd()
    movie_file_path = os.path.join(current_dir, "movie.xlsx")
    links_file_path = os.path.join(current_dir, "links.xlsx")
    batch_data = []  # 用于存储当前批次的数据
    # # 读取已抓取的链接
    # already_visited_urls = set()
    # if os.path.exists(data_file_path):
    #     df = pd.read_excel(data_file_path)
    #     already_visited_urls = set(df['ProductUrl'].dropna())

    img_lists = []
    title_lists = []
    year_lists = []
    time_lists = []
    rank_lists = []
    rate_list = []
    count_list = []
    url_list = []
    id_list = []

    for imgs in tab.eles(".ipc-media ipc-media--poster-27x40 ipc-image-media-ratio--poster-27x40 ipc-media--media-radius ipc-media--base ipc-media--poster-m ipc-poster__poster-image ipc-media__img"):
        img = imgs.ele(".ipc-image").attr('src')
        img_lists.append(img)


    for titles in tab.eles(".ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-a69a4297-2 bqNXEn cli-title with-margin"):
        title = titles.ele(".ipc-title__text").text.split(". ")[1]
        title_lists.append(title)

    for info in tab.eles(".sc-300a8231-6 dBUjvq cli-title-metadata"):
        allinfo = info.eles(".sc-300a8231-7 eaXxft cli-title-metadata-item")
        if len(allinfo) == 2:
            year = allinfo[0].text
            year_lists.append(year)
            time = allinfo[1].text
            time_lists.append(time)
            rank = ''
            rank_lists.append(rank)
        else:
            year = allinfo[0].text
            year_lists.append(year)
            time = allinfo[1].text
            time_lists.append(time)
            rank = allinfo[2].text
            rank_lists.append(rank)


    for rates in tab.eles(".ipc-rating-star--rating"):
        rate = rates.text
        rate_list.append(rate)

    for counts in tab.eles(".ipc-rating-star--voteCount"):
        count = counts.text.replace("(","").replace(")","")
        count_list.append(count)

    for hrefs in tab.eles(".ipc-poster ipc-poster--base ipc-poster--media-radius ipc-poster--dynamic-width ipc-sub-grid-item ipc-sub-grid-item--span-2"):
        url = hrefs.ele(".ipc-lockup-overlay ipc-focusable").attr('href')
        url_list.append(url)
        id = url.replace("https://www.imdb.com/title/","").split("/")[0]
        id_list.append(id)

    return (img_lists,title_lists,year_lists,time_lists,rank_lists,rate_list,count_list,url_list,id_list)

data_file_path = os.path.join(current_dir, "movieList.xlsx")
# data_tuple = (img, title, year, time, rank,rate, count, url, id)
# batch_data.append(data_tuple)
# save_data_to_excel(batch_data, data_file_path)  # 每抓取十条数据保存一次
```

```python
def clickLoadMoreButton(tab,times):
    for i in range(times):
        loadMoreButtons = tab.eles(".ipc-see-more__text")
        loadMoreButtons[0].click()
        print("点击了加载按钮")
        time.sleep(2)
```

```python
def clickSpoilerButton(tab):
    buttons = tab.eles(".ipc-btn ipc-btn--single-padding ipc-btn--center-align-content ipc-btn--default-height ipc-btn--core-base ipc-btn--theme-base ipc-btn--button-radius ipc-btn--on-error ipc-text-button sc-8c7aa573-1 cIpvzQ review-spoiler-button")

    for button in buttons:
        button.click()
        print("点击了剧透按钮")
```

## Step6 读取电影信息表中需要采集评论的电影链接

```python
'''
整体思路是先获取到要采集评论的电影链接，然后逐个访问电影链接，获取评论数据
每个采集100条，点击三次加载更多，然后点击剧透按钮，最终抓取下来所有的评论数据，
每完成1个链接就保存一次数据，避免数据丢失
'''
def getReview(tab,id_lists):
    id_list = []
    star_list = []
    good_list = []
    bad_list = []
    name_list = []
    time_list = []
    title_list = []
    content_list = []
    
    for id in id_lists:
        try:
            review_url = 'https://www.imdb.com/title/' + str(id) + '/reviews/?ref_=tt_ururv_sm'
            print(review_url)
            tab.get(review_url)
            time.sleep(2)
            clickLoadMoreButton(tab,3)
            clickSpoilerButton(tab)

            article_list = tab.eles(".sc-d59f276d-1 euAsTr user-review-item")
            for article in article_list:
                if article.ele(".ipc-rating-star--rating"):
                    star = article.ele(".ipc-rating-star--rating").text
                    star_list.append(star)
                else:
                    star_list.append('')

                id_list.append(id)
                good_list.append(article.ele(".ipc-voting__label__count ipc-voting__label__count--up").text)
                bad_list.append(article.ele(".ipc-voting__label__count ipc-voting__label__count--down").text)
                name_list.append(article.eles(".ipc-link ipc-link--base")[0].text)
                time_list.append(article.ele(".ipc-inline-list__item review-date").text)
                title_list.append(article.ele(".ipc-title__text").text)
                content_list.append(article.ele(".ipc-html-content-inner-div").text)
            print(f"已经抓取了{id}的评论数据")
        except Exception as e:
            print(f"处理电影 {id} 时发生错误: {e}")
            # 尝试刷新页面并重试
            print("尝试刷新页面并重试...")
            tab.refresh()
            time.sleep(random.uniform(1.5, 2.5))
            continue
        

    return (id_list,star_list,good_list,bad_list,name_list,time_list,title_list,content_list)

```

```python
# 初始化 tab1
page = ChromiumPage()    
# 读取或更新商品链接
page.get('https://www.imdb.com/title/tt9362722/reviews/?ref_=tt_ururv_sm')
# 获取第一个标签页对象
tab = page.get_tab()

read_file = pd.read_excel (r'movieList.xlsx')
# 第七个电影的影评只有21条，不需要点击加载更多
id_list = read_file['id'][8:10]

(id_list,star_list,good_list,bad_list,name_list,time_list,title_list,content_list) = getReview(tab,id_list)

# 创建 DataFrame
data = {
    'id': id_list,
    'star': star_list,
    'good': good_list,
    'bad': bad_list,
    'name': name_list,
    'time': time_list,
    'title': title_list,
    'content': content_list,
}
df = pd.DataFrame(data)

# 指定保存路径
output_file = 'reviewList1.xlsx'

# 将 DataFrame 写入 Excel 文件
df.to_excel(output_file, index=False)

print(f"数据已成功写入 {output_file}")
```
